```python
# Prepare your data (e.g. load and tokenize your dataset)
import pandas as pd 
dataset=pd.read_csv('Diabetes.csv')
dataset.head
```

```python
dataset.describe
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.dropna()
dataset.isnull().sum()
```

```python
dataset=dataset.drop(['diag_1_desc','citoglipton','glyburide.metformin','metformin.rosiglitazone','metformin.pioglitazone','glimepiride.pioglitazone','acetohexamide','max_glu_serum','glipizide','glyburide','tolbutamide','tolazamide','pioglitazone','troglitazone','rosiglitazone','miglitol','acarbose','chlorpropamide','glimepiride','A1Cresult','metformin','repaglinide','nateglinide','weight','payer_code','diag_2_desc','diag_3_desc','change','rowID','admission_type_id','discharge_disposition_id','admission_source_id','diag_1','diag_2','diag_3'],axis=1)
```

```python
dataset['race'].value_counts()
```

```python
def clean_race(x):
    if 'Caucasian' in x:
        return 'Caucasian'
    if 'AfricanAmerican' in x:
        return 'AfricanAmerican'
    if 'Hispanic' in x:
        return 'Hispanic'
    if 'Asian' in x:
        return 'Asian'
    return 'Other'
dataset['race']=dataset['race'].apply(clean_race)
dataset
```

```python
dataset['age'].value_counts()
```

```python
def clean_age(x):
    if '[70-80)' in x:
        return '75'
    if '[60-70)' in x:
        return '65'
    if '[50-60)' in x:
        return '55'
    if '[80-90)' in x:
        return '85'
    if '[40-50)' in x:
        return '45'
    if '[30-40)' in x:
        return '35'
    if '[20-30)' in x:
        return '25'
    if '[10-20)' in x:
        return '15'

    return '5'
dataset['age']=dataset['age'].apply(clean_age)
dataset
```

```python
dataset['num_lab_procedures'].value_counts()
```

```python
dataset=dataset.rename({"medical_specialty":"specialty"},axis=1)
```

```python
def clean_medical_specialty(x):
    if '?' in x:
        return 'Non medical'
    return 'Medical'
dataset['specialty']=dataset['specialty'].apply(clean_medical_specialty)
dataset
```

```python
dataset.head
```

```python
dataset['num_lab_procedures'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_race=LabelEncoder()
dataset['race']=l_race.fit_transform(dataset['race'])
l_gender=LabelEncoder()
dataset['gender']=l_gender.fit_transform(dataset['gender'])
l_specialty=LabelEncoder()
dataset['specialty']=l_specialty.fit_transform(dataset['specialty'])
l_examide=LabelEncoder()
dataset['examide']=l_examide.fit_transform(dataset['examide'])
l_insulin=LabelEncoder()
dataset['insulin']=l_insulin.fit_transform(dataset['insulin'])
l_glipizide_metformin=LabelEncoder()
dataset['glipizide.metformin']=l_glipizide_metformin.fit_transform(dataset['glipizide.metformin'])
l_diabetesMed=LabelEncoder()
dataset['diabetesMed']=l_diabetesMed.fit_transform(dataset['diabetesMed'])
l_readmitted=LabelEncoder()
dataset['readmitted']=l_readmitted.fit_transform(dataset['readmitted'])
```

```python
dataset
```

```python
dataset['specialty'].unique
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
X=dataset.drop(['readmitted'],axis=1)
Y=dataset['readmitted']
dataset['race'].value_counts()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=42)
```

```python
model1=RandomForestClassifier(n_estimators=1000)
```

```python
model2=xgb.XGBClassifier()
```

```python
model1.fit(X_train,Y_train)
```

```python
predictions=model1.predict(X_train)
accuracy=accuracy_score(predictions,Y_train)
accuracy
```

```python
predictions1=model1.predict(X_test)
```

```python
accuracy=accuracy_score(predictions1,Y_test)
accuracy
```
